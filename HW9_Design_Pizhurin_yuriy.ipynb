## 2. Написать DAG Airflow для ML-системы для расчета складских запасов сетевого магазина

1) Выберите архитектуру ML-конвейера (из архитектур в [задании 1](#cell-3f512028)).

2) Вспомните раздел Continuous Training Loop из конспекта этого модуля и напишите DAG Airflow для демонстрации концепции непрерывного переобучения (Continuous Training, CT).

3) ML-система должна **автоматически** (без участия человека) принимать решение о начале обучения, если:
* суммарно по всем кассам прошло более 10 млн чеков, и они загружены на S3;
* произошло незапланированное снижение метрики *accuracy < 0.85* из-за массового спроса на клон игрушки Лабубу;
* срок действия модели истекает через 1 час.


*Ожидаемый артефакт: код в [ячейке](#scrollTo=W0FaHkiQsW5J)*


````python
# Выбрана Батч архитектура, потому что:
# - данные о чеках накапливаются партиями
# - переобучение модели запасов не нужно производить мгновенно


from datetime import datetime
from airflow import DAG
from airflow.providers.standard.operators.python import PythonOperator, BranchPythonOperator
from airflow.providers.standard.operators.empty import EmptyOperator


# Функции для проверки условий
def check_s3_data():
    receipts_count = 12_000_000
    print(f"Количество чеков на S3: {receipts_count}")

    return receipts_count > 10_000_000


def check_model_accuracy():
    accuracy = 0.82
    print(f"Текущий accuracy: {accuracy}")

    return accuracy < 0.85


def check_model_expiration():
    hours_before_expiration = 0.5

    print(f"Срок действия модели истекает через: "
          f"{hours_before_expiration} часа")

    return hours_before_expiration < 1


# Функция определяющая необходимость переобучения
def decide_retraining():
    s3_ready = check_s3_data()
    accuracy_problem = check_model_accuracy()
    model_expired = check_model_expiration()

    if s3_ready or accuracy_problem or model_expired:
        print("Необходимо переобучение модели")
        return "train_model"

    print("Переобучение не требуется")
    return "skip_training"



# ML pipeline
def ingest_data():
    print("Loading new data from S3...")


def train_model():
    print("Запуск обучения модели...")


def evaluate_model():
    print("Оценка качества новой модели...")


def register_model():
    print("Регистрация модели в MLflow...")


def deploy_model():
    print("Деплой новой модели в production...")



# Создание DAG
with DAG(
    dag_id="warehouse_stock_ml_pipeline",
    start_date=datetime(2025, 1, 1),
    schedule="@hourly",
    catchup=False,
    default_args={
        "owner": "mlops_engineer",
        "retries": 1,
        "retry_delay": timedelta(minutes=5)
    },
    description="Continuous Training pipeline "
                "для прогнозирования складских запасов"
) as dag:

    decide = BranchPythonOperator(
        task_id="decide_retraining",
        python_callable=decide_retraining
    )

    ingest = PythonOperator(
        task_id="ingest_data",
          python_callable=ingest_data)

    train = PythonOperator(
        task_id="train_model",
        python_callable=train_model
    )

    evaluate = PythonOperator(
        task_id="evaluate_model",
        python_callable=evaluate_model
    )

    register = PythonOperator(
        task_id="register_model",
        python_callable=register_model
    )

    deploy = PythonOperator(
        task_id="deploy_model",
        python_callable=deploy_model
    )

    skip = EmptyOperator(
        task_id="skip_training"
    )


decide >> [ingest, skip]

ingest >> train >> evaluate >> register >> deploy  
````

2) Используя [adr-tools](https://github.com/npryce/adr-tools) запишите причины (по которым мы приняли решение улучшать систему) таким образом, чтобы такое решение мог легко проверить любой сотрудник компании.

*Ожидаемый артефакт: текстовое описание причин для улучшения системы в [ячейке](#scrollTo=YBxl5LWDVmqr)*

````python
import numpy as np
from scipy import stats

np.random.seed(42)

n_hist = 30
accuracy_historical = np.random.normal(loc=0.91, scale=0.02, size=n_hist)

n_new = 14
accuracy_new = np.random.normal(loc=0.82, scale=0.03, size=n_new)

t_stat, p_value = stats.ttest_ind(accuracy_historical, accuracy_new,
                                   equal_var=False)  # Welch's t-test

diff_mean = accuracy_historical.mean() - accuracy_new.mean()
se = np.sqrt(accuracy_historical.std(ddof=1)**2 / n_hist +
             accuracy_new.std(ddof=1)**2 / n_new)
df_welch = (se**2)**2 / (
    (accuracy_historical.std(ddof=1)**2 / n_hist)**2 / (n_hist - 1) +
    (accuracy_new.std(ddof=1)**2 / n_new)**2 / (n_new - 1)
)
t_crit = stats.t.ppf(0.975, df=df_welch)
ci_lower = diff_mean - t_crit * se
ci_upper = diff_mean + t_crit * se

alpha = 0.05
print("=" * 60)
print("Проверка гипотезы: изменилась ли accuracy модели?")
print("=" * 60)
print(f"Историческая accuracy:  {accuracy_historical.mean():.4f} ± {accuracy_historical.std(ddof=1):.4f}")
print(f"Новая accuracy:         {accuracy_new.mean():.4f} ± {accuracy_new.std(ddof=1):.4f}")
print(f"Разность средних:       {diff_mean:.4f}")
print(f"95% ДИ разности:        [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"t-статистика:           {t_stat:.4f}")
print(f"p-value:                {p_value:.6f}")
print()

if p_value < alpha and ci_lower > 0:
    verdict = (
        f"p-value ({p_value:.4f}) < α ({alpha}) и нулевая гипотеза об "
        f"отсутствии изменений ОТВЕРГАЕТСЯ.\n"
        f"95% ДИ разности [{ci_lower:.4f}, {ci_upper:.4f}] не содержит 0.\n"
        f"Принимаю решение: ЗАПУСТИТЬ ВНЕПЛАНОВОЕ ПЕРЕОБУЧЕНИЕ модели.\n"
        f"Падение accuracy ({diff_mean:.3f}) статистически значимо и превышает "
        f"деловой порог 0.06."
    )
else:
    verdict = (
        f"p-value ({p_value:.4f}) ≥ α ({alpha}). Нулевая гипотеза не отвергается.\n"
        f"Принимаю решение: ничего не менять, осознаю риски."
    )

print("Вывод:")
print(verdict)
````

## ADR-001: Запуск внепланового переобучения модели расчёта складских запасов

```
# ADR-001: Запуск внепланового переобучения модели
Date: 2025-05-24
Status: Accepted
Deciders: ML-команда (Пижурин Ю.)
```

### Контекст

Модель прогнозирования складских остатков показала падение accuracy с **0.91** до **0.82**  
за 14 дней после резкого спроса на новую игрушку (аналог Лабубу), изменившего паттерн потребления.  
Двухвыборочный тест Уэлча: **t = 12.3, p < 0.0001**, 95% ДИ разности [+0.074, +0.106].

### Рассмотренные варианты

1. **Ничего не делать** — риск: accuracy продолжит падать, ошибки в заказах товара.  
2. **Ручной запуск переобучения** — задержка ≥ 4 часа, зависит от доступности инженера.  
3. **Автоматический внеплановый ретрейн по алерту (выбрано)** — DAG запускается немедленно при accuracy < 0.85.

### Решение

Поскольку **нулевая гипотеза об отсутствии изменений отвергается** (p < 0.001),  
а падение accuracy статистически значимо и превышает бизнес-порог (Δ > 0.06),  
**принимаем решение: запустить внеплановое переобучение автоматически**.

Триггер `accuracy < 0.85` в `check_triggers()` DAG обеспечивает автономную реакцию  
без участия человека в течение следующего hourly-запуска (≤ 1 час).

### Последствия

- Модель восстановит качество без ручного вмешательства.  
- Решение документировано и воспроизводимо (любой джун может проверить p-value).  
- Внеплановое переобучение потребляет ~15 мин CPU на EC2 — приемлемо.  
- Если дрейф продолжится, нужно рассмотреть переход на онлайн-обучение (ADR-002).

Батч-подход оказался оптимальным: данные о чеках накапливаются партиями и не требуют  
обработки в реальном времени, а переобучение линейной модели занимает считанные минуты.  
DAG Airflow автоматически проверяет триггеры и самостоятельно принимает решение 
о запуске переобучения без участия человека.

IaC-конфигурация на Terraform позволила описать всю инфраструктуру.

Мониторинг SLI и SLO позволяет чётко определить, когда и кто должен вмешаться: 
большинство проблем система решает самостоятельно.

Применен статистический теста Уэлча для принятия решения о переобучении, 
изменение метрики подтверждается расчётом с доверительным интервалом и p-value, 
который может проверить любой сотрудник.
